```python
from huggingface_hub import HfApi
import os

HF_TOKEN    = os.getenv("HF_TOKEN")
repo_id     = "aloofzebra03/VR2_finetuned-paligemma-lora"
adapter_dir = "/kaggle/input/16-may-paligemma-mix/finetuned_paligemma_mix/lora_adapters"

api = HfApi()
api.create_repo(repo_id=repo_id, repo_type="model", exist_ok=True, token=HF_TOKEN)

api.upload_folder(
    folder_path=adapter_dir,
    path_in_repo="",       # root of the repo
    repo_id=repo_id,
    repo_type="model",
    token=HF_TOKEN
)

print(f" Uploaded {adapter_dir} → https://huggingface.co/{repo_id}")
```
